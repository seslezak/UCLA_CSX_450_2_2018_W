```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scores = pd.DataFrame([pd.Series([9,6,7,6,6,9,7,8,8,8,6,8,6,6,10,8,6,7,7,8,8]),
                       pd.Series([7,6,5,8,5,6,3,5,6,5,9,7,6,4,7,7,7,8,6,8,7])]).T
scores.columns = ['quiz_2', 'quiz_3']
```

```python
sns.distplot(scores.quiz_2, label='quiz 2')
sns.distplot(scores.quiz_3, label='quiz 3')
plt.legend()
scores.describe()
```

```python
z_scores = (scores-scores.mean())/scores.std()
adj_scores = np.round(1.5*z_scores + 12)
scores = pd.merge(scores, adj_scores, left_index=True, right_index=True)
scores.columns = ['quiz_2_original', 'quiz_3_original', 'quiz_2_scaled', 'quiz_3_scaled']
```

```python
pd.merge(scores, z_scores, left_index=True, right_index=True)
```
